```python
ratings = df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sb.barplot(x="Aggregate rating",y="Rating Count", data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sb.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',
data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
# Count plot 

sb.countplot(x = "Rating color", data = ratings, palette=['blue','red',
'orange','yellow','green','green'])
```

```python
# Find the country name thats give zero rating.

df_final[df_final["Aggregate rating"] ==0].groupby('Country').size().reset_index()
```

```python
# Find Out Which currency is used by which Country?

df_final[['Country','Currency']].groupby(['Country', 'Currency']).size().reset_index()
```

```python
# Which Country do have online deliveries option 

df_final[df_final["Has Online delivery"] == "Yes"].groupby(['Country', 'Has Online delivery']).size().reset_index()
```
